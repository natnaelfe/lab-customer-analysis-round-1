9. Perform the data cleaning operations mentioned so far in class

- Delete the column education and the number of open complaints from the dataframe.




```python
data = data.drop(columns=['education', 'number_of_open_complaints'])
```

```python
data
```

- Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace(r'%', '') # remove % sign
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(float) # converts the strings into floats
```

```python
data['customer_lifetime_value'] = round(data['customer_lifetime_value'],0) # rounds down to the nearest decimal
```

```python
data.fillna(value=0, inplace=True) # replace NaN with "0"
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(int) # converts the float into an integer
```

```python
data['customer_lifetime_value']  = data['customer_lifetime_value'] *100 # to multiply by 100
```

- Check for duplicate rows in the data and remove if any.

```python
data.drop_duplicates(inplace=True)
```

- Filter out the data for customers who have an income of 0 or less.

```python
print("Count of clients with an income less than 1 = ", data[data['income'] < 1]['income'].count())
```
